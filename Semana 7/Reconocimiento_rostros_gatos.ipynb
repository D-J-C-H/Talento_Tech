```python
faceClassif = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
```

```python
# Cargar la imagen
image = cv2.imread('gatos3.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Detección de rostros de gatos:
faces = faceClassif.detectMultiScale (gray,scaleFactor=1.1, minNeighbors=0, minSize=(30, 30), maxSize=(200,200)) 
print (faces)
```

```python
# Dibujar rectángulos alrededor de los rostros detectados 
for (x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w, y+h), (0,255,0),2)
```

```python
# Cargar las dos imágenes
image1 = cv2.imread('gatos3.jpg')

# Convertir las imágenes de BGR a RGB para matplotlib
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Crear una figura y ejes utilizando subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Mostrar la primera imagen en el primer eje
axes[0].imshow(image1_rgb)
axes[0].axis('off')  # Deshabilitar los ejes

# Mostrar la segunda imagen en el segundo eje
axes[1].imshow(image2_rgb)
axes[1].axis('off')  # Deshabilitar los ejes

# Ajustar el diseño de las subfiguras
plt.tight_layout()

# Mostrar la figura con ambas imágenes
plt.show()
```
