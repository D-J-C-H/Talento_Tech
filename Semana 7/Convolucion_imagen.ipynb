```python

# Cargar la imagen
image = cv2.imread('gatos3.jpg')
```

```python
# Definir el kernel de convolución 
kernel = np.array([[0, -1, 0],
[-1, 5, -1],
[0, -1, 0]])
```

```python
# Aplicar la convolución
convolved_image = cv2.filter2D (image, -5, kernel)
```

```python
# Cargar las dos imágenes
image1 = cv2.imread('gatos3.jpg')

# Convertir las imágenes de BGR a RGB para matplotlib
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(convolved_image, cv2.COLOR_BGR2RGB)

# Crear una figura y ejes utilizando subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Mostrar la primera imagen en el primer eje
axes[0].imshow(image1_rgb)
axes[0].axis('off')  # Deshabilitar los ejes

# Mostrar la segunda imagen en el segundo eje
axes[1].imshow(image2_rgb)
axes[1].axis('off')  # Deshabilitar los ejes

# Ajustar el diseño de las subfiguras
plt.tight_layout()

# Mostrar la figura con ambas imágenes
plt.show()
```

```python
# Mostrar la imagen original y la imagen convolucionada
cv2.imshow('Original Image', image)
cv2.imshow('Convolved Image', convolved_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Guardar la imagen convolucionada en un archivo JPG
cv2.imwrite('convolved_image_gatos3.jpg', convolved_image)
```
